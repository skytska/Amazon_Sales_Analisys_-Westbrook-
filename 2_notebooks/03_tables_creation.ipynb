## Notebook 02: Creating Logical Tables

In this notebook, the raw transactional dataset is split into logical tables
(customers, products, orders, payment_methods).

The goal is to simulate a relational database structure and prepare the data
for loading into PostgreSQL.

## Why split the data into tables?

The original CSV file contains denormalized transactional data.
In real-world analytics projects, such data is usually stored in a relational
database with separate tables for different entities.

By splitting the data into logical tables, we:
- reduce redundancy,
- improve data clarity,
- and make the dataset suitable for SQL-based analysis.

```python
# load the dataset
import pandas as pd
df = pd.read_csv('1_data/raw/amazon_sales.csv')
```

## Create customers table

The customers table is designed to contain one row per customer.

Although the same customer may appear multiple times in the raw dataset
(with different orders or location details), we deduplicate this table
using CustomerID as the primary key to ensure a clean dimension table.


```python
# Customers: one row per customer (deduplicated by CustomerID)
df_customers = (
    df[
        [
            "CustomerID",
            "CustomerName",
            "City",
            "State",
            "Country"
        ]
    ]
    .drop_duplicates(subset="CustomerID")
    .reset_index(drop=True)
)

df_customers.head()
```

## Create products table

The products table is designed as a dimension table with one row per product.

Even though product attributes may appear inconsistent in the raw dataset,
we treat ProductID as the primary key and deduplicate accordingly to ensure
a consistent and reliable products table.


```python
# Products: one row per product (deduplicated by ProductID)
df_products = (
    df[
        [
            "ProductID",
            "ProductName",
            "Category",
            "Brand"
        ]
    ]
    .drop_duplicates(subset="ProductID")
    .reset_index(drop=True)
)

df_products.head()
```

## Create orders table

The orders table contains transactional, order-level data.
It links customers, products, and payment methods together.

```python
# Orders: transactional data (order-level)
df_orders = (
    df[
        [
            "OrderID",
            "OrderDate",
            "CustomerID",
            "ProductID",
            "Quantity",
            "UnitPrice",
            "Discount",
            "Tax",
            "ShippingCost",
            "TotalAmount",
            "OrderStatus",
            "PaymentMethod"
            
        ]
    ]
    .drop_duplicates()
    .reset_index(drop=True)
)

df_orders.head()
```

## Validate created tables

Perform simple checks to ensure that the tables were created correctly
and contain a reasonable number of rows.

```python
print("Customers:", df_customers.shape)
print("Products:", df_products.shape)
print("Orders:", df_orders.shape)
```

## Summary

At this stage, the raw dataset has been successfully split into logical tables
that resemble a relational database structure.
