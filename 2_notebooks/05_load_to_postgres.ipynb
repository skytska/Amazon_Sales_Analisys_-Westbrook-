## Notebook 04: Load DataFrames to PostgreSQL

In this notebook, we load the cleaned logical tables (customers, products, orders) into a PostgreSQL database.

This allows us to perform SQL queries and create visualizations in Tableau
using a relational database.

## Why PostgreSQL?

- PostgreSQL is a widely used relational database.
- Loading data into SQL tables allows us to:
  - perform joins and aggregations efficiently,
  - enforce constraints and relationships,
  - simulate real-world analytics workflows.

```python
# Set working directory
import local_config
from local_config import *
import os
os.chdir(directory_path)
```

```python
# import necessary libraries
import pandas as pd
from sqlalchemy import create_engine, text
```

```python
# Load cleaned datasets
df_customers = pd.read_csv("1_data/processed/customers_clean.csv")
df_products = pd.read_csv("1_data/processed/products_clean.csv")
df_orders = pd.read_csv("1_data/processed/orders_clean.csv")
```

```python
# Database connection parameters
# It's a good practice to store sensitive information like database credentials in a separate configuration file or environment variables.
# Create `local_config.py` based on `local_config_example.py` and add your local database credentials.

from local_config import *

db_user = DB_USER
db_password = DB_PASSWORD
db_host = DB_HOST
db_port = DB_PORT
db_name = DB_NAME
```

```python
# Test connection
engine = create_engine(
    f"postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}"
)
```

```python
# Load data into PostgreSQL

df_customers.to_sql(
    "customers",
    engine,
    schema="public",
    if_exists="append",
    index=False
)

df_products.to_sql(
    "products",
    engine,
    schema="public",
    if_exists="append",
    index=False
)

df_orders.to_sql(
    "orders",
    engine,
    schema="public",
    if_exists="append",
    index=False
)
```

```python
# Verify data load
with engine.connect() as conn:
    tables = ["orders", "customers", "products"]
    for table in tables:
        result = conn.execute(text(f"SELECT COUNT(*) FROM public.{table};"))
        count = result.fetchone()[0]
        print(f"{table}: {count} rows loaded")
```
